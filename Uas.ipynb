```python
# mengimport folium dan geopandas (mengubah geopandas menjadi gpd)

import folium
import geopandas as gpd

# mengambil file yang ada di penyimpanan

gdf = gpd.read_file('Geology+jawa+Tengah.geojson')
gdf.head()

# menampilkan isi file
```

```python
# menampilkan map

gdf.explore()
```

```python
# memilih tabel yang ingin ditampilkan

gdf = gdf[['NM_LEMBAR', 'NAME', 'FORMATION', 'CLASS_LITH', 'Shape_Leng',  'geometry']].fillna(0)
```

```python
# menampilkan map

gdf.explore()
```

```python
# membuat map dan menambahkan panel kontrol lapisan ke objek peta Folium

m = gdf.explore(column="Shape_Leng", name="Shape_Leng")
folium.LayerControl().add_to(m)
m
```

```python
# membuat map dan objek tooltip untuk lapisan GeoJSON pada peta Folium

m = folium.Map(location = [-7.3581841,109.8902028], zoom_start=8)

tooltip = folium.features.GeoJsonTooltip(fields=['Shape_Leng', 'NAME'],
                                        aliases=["Shape Lenght", "NAME"],
                                        labels=True,
                                        stick=False)
folium.GeoJson(gdf, tooltip=tooltip).add_to(m)
folium.LayerControl().add_to(m)
m
```

```python
m = folium.Map(location = [-7.3581841,109.8902028], zoom_start=8)

# membuat peta choropleth dengan menggunakan pustaka Folium.
# Peta choropleth adalah peta yang menunjukkan data numerik dengan menggunakan,
# warna atau pola pada wilayah geografis tertentu.

cpleth = folium.Choropleth(gdf, data=gdf, key_on='feature.properties.NAME',
                          columns=['NAME', 'Shape_Leng'],
                          fill_color="YlGnBu",
                          legend_name="Shape Lenght",
                          name="Geology Jateng")
cpleth.geojson.add_child(tooltip)
cpleth.add_to(m)
folium.LayerControl().add_to(m)
m
```
